# 新型冠状病毒发展趋势Feb-6

本报告跟踪新型冠状病毒子从1月24号开始至今的多项数据，包括新增确诊病例，新增疑似病例等，旨在分析新冠病毒的发展趋势。本报告所引用的数据为官方数据，其中要说明的是，关于全国新增确诊病例，本报告的实际数据可能和当日的官方数据有小小出入，因为是减去了核减后的实际新增病例。比如：官方通报2月4号0-24时，全新增确诊病例3887例，累计确诊24324例（海南省核减1人）。

```python
import pandas as  pd 
import matplotlib.pyplot as plt
%matplotlib inline

from pylab import * 
import matplotlib
matplotlib.rcParams['font.family'] = 'Microsoft YaHei'
mpl.rcParams['font.sans-serif'] = ['Microsoft YaHei'] #更新字体格式
mpl.rcParams['font.size'] = 9
```

```python
yq=pd.read_csv('疫情.csv',encoding='gbk')
yq
```

## 新增确诊病例增长率和疑似病例增长率

```python
fig=plt.figure(figsize=(8,6))
plt.plot(yq[1:14]['日期'],yq[1:14]['全国新增确诊率（新增确诊病例/上一日总确诊人数）'],label='全国确诊病例新增率')
plt.plot(yq[1:14]['日期'],yq[1:14]['全国新增疑似率'],label='全国疑似病例新增率')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.ylabel('比率')
plt.xlabel('日期')
plt.title('全国确诊病例和疑似病例增长率')
plt.show()
```

从上图我们可以看出:  
1.全国确诊病例新增率在不断减小：1月28号最高为0.65,而2月6号已经下降到0.15。可能原因如下：  
* 最有可能发展为确诊病例的是疑似病例，疑似病例新增率的降低，导致确诊病例新增率同步降低  
    
2.全国疑似病例新增率也在不断减小：从峰值1.418（1月27号）下降到0.229（2月6号）。可能原因如下：  
* 病毒的传播可能已经得到一定程度的控制
* 可能只是周期性的表现，因为病毒的潜伏期一般为14天，而我们统计的数据也只是13天，所以可能只是一个周期内的固有表现，因此也不可疏忽大意。  

3.确诊病例新增率和疑似病例新增率之间的差值不断减小。可能的原因如下：  
* 对疑似特征的预判已经比较准确，因为如何疑似病例很多，而新增病例很少，则说明大部分疑似病例都不会是确诊病例。



## 死亡病例增长率和治愈病例增长率

```python
fig=plt.figure(figsize=(8,6))
plt.plot(yq[1:14]['日期'],yq[1:14]['新增死亡率（新增死亡病例/上一日总确诊人数*）'],c='red',label='新增死亡率')
plt.plot(yq[1:14]['日期'],yq[1:14]['新增治愈率（新增治愈病例/上一日总确诊人数）'],c='green',label='新增治愈率')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.ylabel('比率')
plt.xlabel('日期')
plt.title('全国死亡病例和治愈病例增长率')
plt.show()
```

从上图我们可以看到：  
 1.死亡病例增长率不断下降，从1月25号的峰值的0.0193下降到2月6号的0.003  
 2.治愈病例增长率在不断上升，从1月27号的最低值0.001上升到2月6号的0.011
 呈现以上趋势的可能原因如下：
 * 虽然目前还没有特效药，但在维持病人生存方面已经有了不小的进步。
 * 近期新增的大量确诊病例数目稀释了死亡病例增长率。

## 结论

在过去两周，虽然新增病例的绝对人数都维持在高位，但新型冠状病毒的新增确诊和疑似病例增长率都在显著下降，这可能是疫情得到了一定程度的控制，也可能只是一种周期性的表现，因为毕竟病毒的潜伏期有14天左右，是存在一个时间延迟。所以对病毒仍不可掉以轻心。
